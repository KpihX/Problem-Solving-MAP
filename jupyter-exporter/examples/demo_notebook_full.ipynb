## 3Ô∏è‚É£ Masquage Partiel du Code

Le tag `partial_code` affiche uniquement les **2 premi√®res et 2 derni√®res lignes** (configurable).
Le reste est remplac√© par `# ... [N lignes masqu√©es] ...`

````python
# Cellule longue avec du code m√©tier complexe
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Simulation d'une longue cha√Æne de traitement
data = pd.DataFrame({
    'date': pd.date_range('2025-01-01', periods=100),
    'revenue': np.random.normal(50000, 10000, 100),
    'costs': np.random.normal(30000, 5000, 100),
})

# Calculs interm√©diaires
data['profit'] = data['revenue'] - data['costs']
data['margin'] = data['profit'] / data['revenue'] * 100

# Agr√©gation par semaine
weekly = data.resample('W', on='date').mean()

# Validation des r√©sultats
assert len(weekly) > 0, "Erreur : donn√©es vides"
assert weekly['margin'].mean() > 0, "Erreur : marges n√©gatives"

print(f"üîç Analyse termin√©e : {len(data.columns)} variables trait√©es")
print("‚úÖ Validation r√©ussie")
````
